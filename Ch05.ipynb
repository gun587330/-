```python
# 함수와 메서드 차이

# 내장 함수 : 기본 함수 형태로, 함수 이름과 괄호를 입력하여 사용. 파이썬에 내장되어 있어 별도 패키지 설치 없이 이용 가능.
# EX) sum(var), max(var) 

# 패키지 함수 : 패키지를 먼저 로드(import)해야 사용 가능. package.function() 형태로 사용함.
# EX) import pandas as pd / pd.read_csv('exam.csv'), import pandas as pd / pd.DataFrame({'x':[1, 2, 3]})

# 메서드(method) : 변수가 지니고 있는 함수. 변수명을 입력한 다음 점 찍고 메서드 이름과 괄호 입력하여 사용
# EX) df=pd.read_csv('exam.csv') /. df.head()
# > 메서드는 모든 변수에 사용할 수 있는 것이 아님.
```

```python
# 어트리뷰트(attribute) : 변수가 지니고 있는 값. '변수명 뒤에 점을 찍고 어트리뷰트 이름' 형태로 출력
# 메서드와 같이 변수명 뒤에 점을 찍고 입력하지만, 메서드와 달리 괄호는 사용하지 않음.
# EX) df.shape : o , list.shape : x
# > 어트리뷰트는 모든 변수에 사용할 수 있는 것이 아님.
```

```python
# 05-3 파생변수 만들기
# 파생변수 : 기존의 변수를 변형해 만든 변수
df = pd.DataFrame({'var1' : [4, 3, 8], # 데이터 프레임 df 생성
                   'var2' : [2, 6, 1]})
# var1과 var2변수의 값을 더한 var_sum이라는 파생변수를 만들어 df에 추가
df['var_sum'] = df['var1'] + df['var2'] # var_sum 파생변수 만들기
# var1과 var2변수의 값을 더한 다음 2로 나눠 var_mean이라는 파생변수를 만들어 df에 추가
df['var_mean'] = (df['var1'] + df['var2'])/2
df
```

```python
# mpg 통합 연비 변수 만들기
mpg = pd.read_csv('mpg.csv') # mpg데이터 불러오기
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2 # 통합 연비 변수 만들기
mpg
```

```python
sum(mpg['total'])/len(mpg) # total 합계를 행 수로 나누기
```

```python
mpg['total'].mean() # sum(), len() 대신  df.mean() 메서드를 이용하면 평균을 더 간단하게 구할 수 있음
```

```python
# 조건문을 활용해 파생변수 만들기
# 조건에 따라 서로 다른 값을 반환하는 '조건문 함수'러 파생변수 만들기
# 1. 기준값 정하기
mpg['total'].describe() # 요약 통계량 출력
mpg['total'].plot.hist() # total변수를 히스토그램을 통해 파악하기

# 히스토그램을 보고 알 수 있는 사항
# > total 연비의 평균과 중앙값이 약 20이다.
# > total 연비가 20~22.5 사이인 자동차 모델이 가장 많다.
# > total 연비는 대부분 25 이하이고, 25를 넘기는 자동차 모델은 많지 않다.
```

```python
# 2. 합격 판정 변수 만들기
# 조건문(conditional statements) 함수 : 조건에 따라 서로 다른 값을 반환하는 함수
# Numpy : 배열 연산, 통계치 계산 등 수치 연산에 자주 사용하는 패키지
# np.where(조건, 조건에 맞을 때 부여할 값, 조건에 맞지 않을 때 부여할 값)
import numpy as np # numpy 패키지를 np로 로드(load)
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail') # 20이상이면 pass, 그렇지 않으면 fail 부여
mpg
```

```python
# 3. 빈도표(Frenquency table)로 합격 판정 자동차 수 살펴보기
# 합격 판정을 받은 자동차와 불합격 판정을 받은 자동차가 몇 대인지 알아보기
mpg['test'].value_counts() # 연비 합격 빈도표 만들기
```

```python
# 4. 막대 그래프로 빈도 표현하기
count_test = mpg['test'].value_counts() # 연비 합격 빈도표를 변수에 할당
count_test.plot.bar() # 연비 합격 빈도 막대 그래프 만들기
```

```python
# 위 출력한 그래프를 보면 축 이름이 회전되어 있음. 이는 df.plot.bar()의 기본값이 축 이름을 90도 회전하게 하는 것이기 때문
count_test.plot.bar(rot = 0) # rot : 회전 각도
```

```python
# 중첩 조건문 활용하기
# 1. 연비 등급 변수 만들기
# total 기준으로 A, B, C 등급 부여
# 범주 값을 3가지 이상으로 부여하려면 np.where()안에 np.where()을 중첩하여 나타내야 함.
mpg['grade'] = np.where(mpg['total'] >= 30, 'A',
                np.where(mpg['total'] >= 20, 'B', 'C'))
mpg.head() # 앞 5개만 간략히 확인하여 데이터 확인
```

```python
# 2. 빈도표와 막대 그래프로 연비 등급 살펴보기
count_grade = mpg['grade'].value_counts() # 등급 빈도표 만들기
count_grade
```

```python
count_grade.plot.bar(rot = 0) # 등급 빈도 막대 그래프 만들기 > 빈도 기준으로 내림차순 정리되어 있음(디폴트)
```

```python
# 알파벳순으로 막대 정렬하기
count_grade = mpg['grade'].value_counts().sort_index() #메서드 체이닝(method chaining) 변수에 여러 메서드를 순서대로 적용할 수 있어 출력 결과를 변수에 할당하고 다시 불러오는 작업을 반복하지 않아도 됨.(숏코딩)
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```

```python
# 분석 도전
# midwest.csv는 미국 동북중부(East North Central States) 437개 지역의 인구통계 정보를 담고 있습니다. midwest.csv를 이용해 데이터 분석 문제를 해결해 보세요.
midwest = pd.read_csv('midwest.csv') # midwest.csv 파일 불러와서 midwest에 지정
```

```python
# 문제 1 midwest.csv를 불러와 데이터의 특징을 파악하세요.
midwest.head() # 앞부분 출력
midwest.tail() # 뒷부분 출력
midwest.shape # 행/열 개수 출력
midwest.info() # 변수 속성 출력
midwest.describe() # 요약 통계랑 출력
```

```python
# 문제 2 poptotal(전체 인구) 변수를 total로, popasian(아시아 인구) 변수를 asian으로 수정하세요.
midwest_new = midwest.copy() # 데이터를 잃어버릴 수 있으므로 복사본 사용
midwest_new = midwest_new.rename(columns = {'poptotal' : 'total'}) # poptotal > total
midwest_new = midwest_new.rename(columns = {'popasian' : 'asian'}) # popasian > asian
midwest_new
```

```python
# 문제 3 total, asian 변수를 이용해 '전체 인구 대비 아시아 인구 백분율' 파생변수를 추가하고, 히스토그램을 만들어 분포를 살펴보세요.
midwest_new['per_asian'] = midwest_new['asian']/midwest_new['total']*100 # 백분율 변수 추가 
midwest_new['per_asian'].plot.hist() # per_asian 빈도수 히스토그램 
# ratio : 비율 이라는 뜻이 있으므로 답지처럼 ratio로 쓰는 것이 더 좋을 듯
```

```python
# 문제 4 아시아 인구 백분율 전체 평균을 구하고, 평균을 초과하면 'large', 그 외에는 'small'을 부여한 파생변수를 만들어 보세요.
avg = sum(midwest_new['per_asian'])/len(midwest_new['per_asian']) # per_asian 평균
# 답지 : midwest['ratio'].mean()

midwest_new['test'] = np.where(midwest_new['per_asian'] > avg, 'large', 'small') # large, small 부여
midwest_new
```

```python
# 문제 5 'large'와 'small'에 해당하는 지역이 얼마나 많은지 빈도표와 빈도 막대 그래프를 만들어 확인해 보세요.
midwest_new['test'].value_counts()
```

```python
count_test = midwest_new['test'].value_counts()
count_test.plot.bar(rot = 0)
```
