```python
# 06-5 파생변수 추가하기
# df.assign(새 변수 = 변수 만드는 공식)
# 새로 만들 변수명에는 따옴표를 입력하지 않음
exam. assign(total = exam['math'] + exam['english'] + exam['science'])
```

```python
# 여러 파생 변수 한 번에 추가하기
exam. assign(
    total = exam['math'] + exam['english'] + exam['science'], # total 추가
    mean = (exam['math'] + exam['english'] + exam['science'])/3) # mean 추가
```

```python
# df.assign()에 np.where() 적용하기
# df.assigndp Numpy 패키지의 where()를 적용하면 조건에 따라 다른 값을 부여한 변수를 추가할 수 있음.
exam.assign(test = np.where(exam['science'] >= 60, 'pass', 'fail')).head()
```

```python
# total 변수 추가, total 기준 정렬
# 변수를 추가하고 나면 이어지는 pandas 함수에 바로 활용할 수 있음.
exam.assign(total = exam['math'] + exam['english'] + exam['science'])\
                    .sort_values('total').head()
```

```python
# lambda 이용해 dataFrame 명 줄여 쓰기
# df.assign을 사용할 때 변수명 앞에 데이터 프레임명을 반복해서 입력해야 하므로 번거로움
# lambda x: 는 데이터 프레임명 자리에 x를 입력하겠다는 의미

# 긴 데이터 프레임명 지정
long_name = pd.read_csv('exam.csv')

# long_name 직접 입력
long_name.assign(new = long_name['math'] 
                     + long_name['english'] 
                     + long_name['science'])

# long_name 대신 x 입력
long_name.assign(new = lambda x: x['math'] + x['english'] + x['science']).head()
```

```python
# 앞에서 만든 변수를 활용해 다시 변수 만들기
# 앞에서 만든 파생변수를 이용해 다시 파생변수를 만들 때는 lambda를 이용해 df명을 약어로 입력해야 함
exam. assign(
    total = exam['math'] + exam['english'] + exam['science'],
    mean = lambda x: (x['total'])/3)
# > 서로 다른 문자로 입력하여 읽기 불편
exam. assign(
    total = lambda x: x['math'] + x['english'] + x['science'],
    mean  = lambda x: x['total']/3).head()
# > 데이터 명을 통일해야 깔끔함.
```

```python
# 혼자서 해보기
# Q1 mpg 데이터 복사본을 만들고, cty와 hwy를 더한 '합산 연비 변수'를 추가하세요.
mpg = pd.read_csv('mpg.csv')
mpg_copy = mpg.assign( total = mpg['cty'] + mpg['hwy'])
mpg_copy
```

```python
# Q2 앞에서 만든 '합산 연비 변수'를 2로 나눠 '평균 연비 변수'를 추가하세요.
mpg_copy = mpg_copy.assign( mean = lambda x: x['total'] / 2 )
mpg_copy
```

```python
# Q3 '평균 연비 변수'가 가장 높은 자동차 3종의 데이터를 출력하세요.
mpg_copy.sort_values('mean').head(3)
```

```python
# Q4 1~3번 문제를 해결할 수 있는 하나로 연결된 pandas 구문을 만들어 실행해 보세요. 
# 데이터는  복사본 대신 mpg 원본을 이용하세요.
mpg.assign(total = lambda x: x['cty'] + x['hwy'],
           mean  = lambda x: x['total'] / 2)\
           .sort_values('mean').head(3)
```

```python
# 06-6 집단별로 요약하기
# '집단별 평균', '집단별 빈도'처럼 각 집단을 요약한 값을 구할 때 df.groupby()와 df.agg()를 사용
# df.agg(요약값을 할당할 변수 = ('요약할 때 사용할 변수', '요약할 때 사용할 함수 > 괄호 사용 x'))

# math 평균 구하기
exam = pd.read_csv('exam.csv')
exam.agg(mean_math = ('math', 'mean'))
```

```python
# 집단별 요약 통계량 구하기
# nclass별로 분리하기
# df.groupby() > variable을 index로 바꿈
exam.groupby('nclass') \
    .agg(mean_math = ('math', 'mean'))
```

```python
# df.groupby()에 as_index = False 입력하면 variable을 index로 바꾸지 않음.
exam.groupby('nclass', as_index = False) \
    .agg(mean_math = ('math', 'mean'))
```

```python
# df.assign()처럼 여러 변수를 한 번에 추가하는 것 가능
exam.groupby('nclass') \
    .agg(mean_math   = ('math', 'mean'),
         sum_math    = ('math', 'sum'),
         median_math = ('math', 'median'),
         n           = ('nclass', 'count'))
```

```python
# agg()에 자주 사용하는 요약 통계량 함수
# mean() 평균
# std() 표준편차
# sum() 합계
# median() 중앙값
# min() 최소값
# max() 최대값
# count() 빈도(개수)
```

```python
# 제조 회사 및 구동 방식별 분리
# cty 평균 구하기
mpg.groupby(['manufacturer', 'drv'])\
   .agg(mean_cty = ('cty', 'mean'))  
```

```python
# 집단별로 다시 집단 나누기
mpg.groupby(['manufacturer', 'drv']) \
   .agg(mean_cty = ('cty', 'mean'))
```

```python
mpg.query('manufacturer == "audi"') \
   .groupby(['drv']) \
   .agg(n = ('drv', 'count'))
```

```python
exam.groupby('id').agg(mean_math = ('math', 'mean'))
```

```python
# pandas 함수 조합하기

# 제조 회사별로 'suv'자동차의 
# 도시 및 고속도로 합산 연비 평균을 
# 구해 내림차순으로 정렬하고, 1~5위까지 출력하기

# 코드를 작성하기 전에 어떤 절차로 어떤 함수를 사용할지 생각하면서 정리
# 1 suv 추출 > query()
# 2 합산 연비 변수 만들기 > assign()
# 3 회사별로 분리 > groupby()
# 4 합산 연비 평균 구하기 > agg()
# 5 내렴차순 정렬 > sort_values()
# 6 1~5위까지 출력 > head()

mpg.query('category == "suv"') \
   .assign(total = (mpg['hwy'] + mpg['cty']) / 2) \
   .groupby('manufacturer') \
   .agg(mean_tot = ('total', 'mean')) \
   .sort_values('mean_tot', ascending = False) \
   .head()
```

```python
# 혼자서 해보기
# Q1
mpg.groupby('category').agg(mean_cat = ('cty', 'mean'))
```

```python
# Q2
mpg.groupby('category') \
   .agg(mean_cat = ('cty', 'mean')) \
   .sort_values('mean_cat', ascending = False)
```

```python
# Q3
mpg.groupby('manufacturer') \
   .agg(mean_hwy =('hwy', 'mean')) \
   .head(3)
```

```python
# Q4
mpg.query('category == "compact"') \
   .groupby('manufacturer') \
   .agg(totle = ('manufacturer', 'count')) \
   .sort_values('totle', ascending = False )
```

```python
# 가로로 합치기
# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id'     : [1, 2, 3, 4, 5],
                      'midterm': [60, 80, 70, 90, 85]})
# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id'     : [1, 2, 3, 4, 5],
                      'final': [70, 83, 65, 95, 80]})
```

```python
# 데이터를 가로로 합칠 땐, pd.merge() 이용
# 1. pd.merge()에 결합할 데이터 프레임명 나열
# 2. how = 'left' 를 이용하여 오른쪽에 입력한 DF을 왼쪽 DF에 결합
# 3. 기준으로 삼을 변수명을 on에 입력
total = pd.merge(test1, test2, how = 'left', on = 'id')
total
```

```python
# 다른 데이터를 활용해 변수 추가하기
exam = pd.read_csv('exam.csv')
name = pd.DataFrame({'nclass'   : [1, 2, 3, 4, 5],
                     'teacher'  : ['kim', 'lee', 'park', 'choi', 'jung']})
name
```

```python
# # 세로로 합치기
# 데이터를 세로로 합칠 땐 pd.concat() 를 이용
# pd.concat()에 결합할 DF 명을 []로 나열하면 됨
# Data를 세로로 합칠 땐 변수명이 같아야 함
```

```python
# 학생 1-5번 시험 데이터 만들기
group_a = pd.DataFrame({'id'   : [1, 2, 3, 4, 5],
                        'test' : [60, 80, 70, 90, 85]})
# 학생 6-10번 시험 데이터 만들기
group_b = pd.DataFrame({'id'   : [6, 7, 8, 9, 10],
                        'test' : [70, 83, 65, 95, 80]})
```

```python
# 데이터 합쳐서 group_all에 할당
group_all = pd.concat([group_a, group_b])
group_all
# index 0~4가 중복됨. > 새로 부여하기 위해선 pd.concat()에 ignore_index = True를 입력
```
