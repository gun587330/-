```python
# 05-3 파생변수 만들기
# 파생변수 : 기존의 변수를 변형해 만든 변수
import pandas as pd
df = pd.DataFrame({'var1' : [4, 3, 8], # 데이터 프레임 df 생성
                   'var2' : [2, 6, 1]})
# var1과 var2변수의 값을 더한 var_sum이라는 파생변수를 만들어 df에 추가
df['var_sum'] = df['var1'] + df['var2'] # var_sum 파생변수 만들기
# var1과 var2변수의 값을 더한 다음 2로 나눠 var_mean이라는 파생변수를 만들어 df에 추가
df['var_mean'] = (df['var1'] + df['var2'])/2
df
```

```python
# mpg 통합 연비 변수 만들기
import pandas as pd
mpg = pd.read_csv('mpg.csv') # mpg데이터 불러오기
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2 # 통합 연비 변수 만들기
mpg
```

```python
sum(mpg['total'])/len(mpg) # total 합계를 행 수로 나누기
```

```python
mpg['total'].mean() # sum(), len() 대신  df.mean() 메서드를 이용하면 평균을 더 간단하게 구할 수 있음
```

```python
# 조건문을 활용해 파생변수 만들기
# 조건에 따라 서로 다른 값을 반환하는 '조건문 함수'러 파생변수 만들기
# 1. 기준값 정하기
mpg['total'].describe() # 요약 통계량 출력
mpg['total'].plot.hist() # total변수를 히스토그램을 통해 파악하기

# 히스토그램을 보고 알 수 있는 사항
# > total 연비의 평균과 중앙값이 약 20이다.
# > total 연비가 20~22.5 사이인 자동차 모델이 가장 많다.
# > total 연비는 대부분 25 이하이고, 25를 넘기는 자동차 모델은 많지 않다.
```

```python
# 2. 합격 판정 변수 만들기
# 조건문(conditional statements) 함수 : 조건에 따라 서로 다른 값을 반환하는 함수
# Numpy : 배열 연산, 통계치 계산 등 수치 연산에 자주 사용하는 패키지
# np.where(조건, 조건에 맞을 때 부여할 값, 조건에 맞지 않을 때 부여할 값)
import numpy as np # numpy 패키지를 np로 로드(load)
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail') # 20이상이면 pass, 그렇지 않으면 fail 부여
mpg
```

```python
# 3. 빈도표(Frenquency table)로 합격 판정 자동차 수 살펴보기
# 합격 판정을 받은 자동차와 불합격 판정을 받은 자동차가 몇 대인지 알아보기
mpg['test'].value_counts() # 연비 합격 빈도표 만들기
```

```python
# 4. 막대 그래프로 빈도 표현하기
count_test = mpg['test'].value_counts() # 연비 합격 빈도표를 변수에 할당
count_test.plot.bar() # 연비 합격 빈도 막대 그래프 만들기
```

```python
# 위 출력한 그래프를 보면 축 이름이 회전되어 있음. 이는 df.plot.bar()의 기본값이 축 이름을 90도 회전하게 하는 것이기 때문
count_test.plot.bar(rot = 0) # rot : 회전 각도
```
