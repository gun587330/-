```python
# 함수와 메서드 차이

# 내장 함수 : 기본 함수 형태로, 함수 이름과 괄호를 입력하여 사용. 파이썬에 내장되어 있어 별도 패키지 설치 없이 이용 가능.
# EX) sum(var), max(var) 

# 패키지 함수 : 패키지를 먼저 로드(import)해야 사용 가능. package.function() 형태로 사용함.
# EX) import pandas as pd / pd.read_csv('exam.csv'), import pandas as pd / pd.DataFrame({'x':[1, 2, 3]})

# 메서드(method) : 변수가 지니고 있는 함수. 변수명을 입력한 다음 점 찍고 메서드 이름과 괄호 입력하여 사용
# EX) df=pd.read_csv('exam.csv') /. df.head()
# > 메서드는 모든 변수에 사용할 수 있는 것이 아님.
```

```python
# 어트리뷰트(attribute) : 변수가 지니고 있는 값. '변수명 뒤에 점을 찍고 어트리뷰트 이름' 형태로 출력
# 메서드와 같이 변수명 뒤에 점을 찍고 입력하지만, 메서드와 달리 괄호는 사용하지 않음.
# EX) df.shape : o , list.shape : x
# > 어트리뷰트는 모든 변수에 사용할 수 있는 것이 아님.
```
